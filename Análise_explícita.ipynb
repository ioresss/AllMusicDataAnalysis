```python
#google colab
from google.colab import drive
#Numpy
import numpy as np
#Pandas
import pandas as pd
#Pyplot
import matplotlib.pyplot as plt
#seaborn
import seaborn as sns

drive.mount('/drive')
%cd ..
```

```python
df = pd.read_json("/drive/MyDrive/data (2).json")
df
```

#Quais são os humores mais relacionados a cada gênero?

```python
generos = [genero for sublista in df['Genero'] for genero in sublista]
generos = np.unique(generos).tolist()
print("Todos os generos:")
generos
```

```python
humorgenero = {}

for genero in generos:
  df_crop = df[[genero in sublista for sublista in df['Genero']]]
  humores = [humor for sublista in df_crop['Humores'] for humor in sublista]
  humorgenero[genero] = humores
```

```python
len(humorgenero)
```

```python
from wordcloud import WordCloud, STOPWORDS


fig, ax = plt.subplots(figsize=(20,12))
cont = 1
for genero in humorgenero:
  humor_str = ' '.join(humorgenero[genero])
  cloud = WordCloud(background_color="black", width=800, height=400).generate(humor_str)
  plt.subplot(4, 5, cont)
  plt.imshow(cloud); 
  plt.title(genero)
  cont = cont + 1
  ax.set_axis_off() 
```
